# 1. 비정상 데이터를 정상 데이터로 만들기

```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# air passengers 데이터 불러오기
ap = pd.read_csv('../../data/AirPassengers.csv')
```

```python
# 정상데이터로 변환 시 Month데이터는 불필요하므로 제거
ap.drop('Month', axis=1, inplace=True)
```

```python
# column name 앞에 #은 불필요해 보이므로 제거
ap.rename({'#Passengers': 'passengers'}, axis=1, inplace=True)
```

```python
ap.head()
```

```python
# 데이터 시각화
sns.lineplot(data=ap, x=ap.index,  y='passengers')
plt.show()
```

## 분산 일정하게 만들기 (로그 변환)

```python
# log transformation을 통해 분산 일정하게 만들기
log_transformed = np.log(ap)
sns.lineplot(x=log_transformed.index, y=log_transformed['passengers'])
plt.show()
```

- log transformation 후 분산이 일정해진 모습 확인 가능

## 평균 일정하게 만들기 (차분)

```python
# 1차 차분 후 데이터 시각화
diff1 = log_transformed.diff().dropna()
sns.lineplot(x=diff1.index, y=diff1['passengers'])
plt.show()
```

- 1차 차분 후 장기적으로 증가하던 추세가 사라진 모습 확인 가능

## 계절성 제거하기 (계절 차분)

```python
ap_mon = pd.read_csv('../../data/AirPassengers.csv')
ap_mon['Month'] = pd.to_datetime(ap_mon['Month']) # datetime으로 자료형 변환
ap_mon['Year'] = ap_mon['Month'].dt.year # year 추출
ap_mon.rename({'#Passengers': 'passengers'}, axis=1, inplace=True) # 위와 동일하게 컬럼명 변환
```

```python
plt.figure(figsize=(15, 12))

sns.lineplot(data=ap_mon, x='Month', y='passengers', hue='Year', palette='tab20') # 월별 데이터이나 연도별로 색상 구분
plt.gca().xaxis.set_major_locator(mdates.YearLocator())  # 연도 단위로 major tick
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # 연도만 표시

plt.title('Number of Passengers per Month')
plt.xticks(rotation=45)  # 라벨 겹침 방지
plt.show()
```

- 기존 데이터에서 1년 주기로 계절성 띄는 것 확인 가능

```python
seasonally_diffed = diff1.diff(12) #  월별 데이터이고, 1년 단위로 계절성을 띄므로 주기롤 12로 설정 하여 차분
seasonally_diffed.dropna(axis=0, inplace=True) # 결측치 제거
```

```python
sns.lineplot(x=seasonally_diffed.index, y=seasonally_diffed['passengers'])
plt.show()
```

- 계절성 제거 후 데이터 시각화
- 그러나 그래프만 봐서는 정상 데이터인지 확인 어려움 => 검정 시행

## 검정을 통해 정상성 확인하기

```python
from statsmodels.tsa.stattools import adfuller
```

```python
# 검정에 p-value만 확인하기 위해 함수 생성
def adf_test(x):
    stat, p_value, lags, nobs, crit, icb = adfuller(x)
    print(f'검정통계량: {stat}')
    print(f'P-value: {p_value}')
```

```python
adf_test(seasonally_diffed)
```

- p-value가 0.00002 이므로 유의수준 0.05, 0.01 모두에서도 기각 가능 => 단위근이 존재한다는 귀무가설을 기각하고 대립가설인 시계열이 **정상성**을 띈다는 것을 채택 가능

```python
# 분산만 일정하게 만든 경우
adf_test(log_transformed)
```

```python
# 분산, 평균 모두 일정하게 만든 경우
adf_test(diff1)
```

- 로그변환만 진행한 경우, 로그변환 및 1차 차분까지 진행한 경우 모두 귀무가설을 기각하기는 어려워보임 => 계절성까지 제거해야 확실하게 정상데이터로 고려 가능

```python
# id기준으로 데이터 분리하기 위해 train_test_split 대신 사용자 정의 함수 생성
def custom_classification_split(x, y, test_size=0.3):
    num_true = int(y.sum() * test_size)
    num_false = int((y.shape[0] - y.sum()) * test_size)
    test_id = (y[y==True].head(num_true) + y[y==False].head(num_false)).index

    X_train = x.loc[~x['id'].isin(test_id)]
    X_test = x.loc[x['id'].isin(test_id)]
    y_train = y.drop(test_id)
    y_test = y.loc[test_id]
        
    return X_train, X_test, y_train, y_test
```

```python
# 함수를 이용해 train, test 데이터 분리
X_train, X_test, y_train, y_test = custom_classification_split(timeseries, y, test_size=0.2)
```
